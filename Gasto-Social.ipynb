```python
#%% Carga de catálogos
catalogo = pd.read_csv('catalogo_2020.09.07.csv',encoding='ISO-8859-1').fillna('   ')
# Cargamos la base a calificar
#%% Cargar base a calificar
base = pd.read_excel('Base_a_calificar_FINAL.xlsx',encoding='UTF-8').fillna("")
#%% Cargamos los criterios de inclusión 
crmp_inclusion=pd.read_excel('Criterios_Inclusión_Exclusión_Listado_2020.09.07_VF.xlsx', sheet_name='Inclusión').fillna("")
#%% Cargamos los criterios de exclusión
crmp_exclusion=pd.read_excel('Criterios_Inclusión_Exclusión_Listado_2020.09.07_VF.xlsx', sheet_name='Exclusión').fillna("")['CRMP'].tolist()
#%% Cargamos los criterios de inclusión de contraloría
crmp_contraloria=pd.read_excel('Criterios_Inclusión_Exclusión_Listado_2020.09.07_VF.xlsx', sheet_name='Inclusión_Contraloría_social').fillna("")
#%% Cargamos los ramos de los programas que serán calificados con el objetivo 4 
ramos_obj_4=pd.read_excel('Criterios_Inclusión_Exclusión_Listado_2020.09.07_VF.xlsx', sheet_name='Ramo').fillna("")
#%% Cargamos las palabras que entrarán en el stemming
excepciones_stemming=pd.read_csv('excep_stemming.csv').fillna("")['stemming_data'].tolist()
```

```python
# Creamos una función para generar un diccionario a partir de columnas de un dataframe
def creaCategoriasDic(catalogo,*argv):
    n=len(argv)
    agrega=""
    for i in range(n):
        if i !=(n-1):
            agrega=agrega+catalogo[argv[i]].map(str)+","
        else:
            agrega=agrega+catalogo[argv[i]].map(str)
    cat_dic=agrega.apply(lambda x: x.split(",")).to_dict()

    return cat_dic
```

```python
# Creamos el diccionario de categorías de los objetivos 1,2 y 3 
cat_dic_1_3=creaCategoriasDic(catalogo[(catalogo['ID_OBJETIVO']!=5) & (catalogo['ID_OBJETIVO']!=4)]  ,'Categoria_Principal','Categoria_Auxiliar_1','Categoria_Auxiliar_2','Categoria_Auxiliar_3')
# Creamos el diccionario de categorias del objetivo 4
cat_dic_4=creaCategoriasDic(catalogo[catalogo['ID_OBJETIVO']==4]  ,'Categoria_Principal','Categoria_Auxiliar_1','Categoria_Auxiliar_2','Categoria_Auxiliar_3')
# Creamos el diccionario de categorias del objetivo 5
cat_dic_5=creaCategoriasDic(catalogo[catalogo['ID_OBJETIVO']==5]  ,'Categoria_Principal','Categoria_Auxiliar_1','Categoria_Auxiliar_2','Categoria_Auxiliar_3')
# Creamos el diccionario de los id de las categorias de los objetivos 1,2 y 3 
id_cat_dic_1_3=creaCategoriasDic(catalogo[(catalogo['ID_OBJETIVO']!=5) & (catalogo['ID_OBJETIVO']!=4)] ,'id')
# Creamos el diccionario de los id de las categorias del objetivo 4
id_cat_dic_4=creaCategoriasDic(catalogo[catalogo['ID_OBJETIVO']==4] ,'id')
# Creamos el diccionario de los id de las categorias del objetivo 5
id_cat_dic_5=creaCategoriasDic(catalogo[catalogo['ID_OBJETIVO']==5] ,'id')

# Creamos un diccionario de subobjetivos de los objetivos 1,2 y 3
subobjetivo_dic_1_3=catalogo[(catalogo['ID_OBJETIVO']!=5) & (catalogo['ID_OBJETIVO']!=4)]['SUBOBJETIVO'].apply(lambda x: [x]).to_dict()
# Creamos un diccionario de subobjetivos del objetivo 4
subobjetivo_dic_4=catalogo[catalogo['ID_OBJETIVO']==4]['SUBOBJETIVO'].apply(lambda x: [x]).to_dict()
# Creamos un diccionario de subobjetivos del objetivo 5
subobjetivo_dic_5=catalogo[catalogo['ID_OBJETIVO']==5]['SUBOBJETIVO'].apply(lambda x: [x]).to_dict()

# Creamos una lista de los ramos
ramos_obj_4_list=list(ramos_obj_4['Ramo'])
# Creamos un diccionario de los criterios de inclusión
crmp_inclusion_dict={}

for i in range(len(crmp_inclusion)):
    crmp_inclusion_dict.update({crmp_inclusion.iloc[i,0]: crmp_inclusion.iloc[i,2].split("/")})

# Creamos un diccionario de los criterios de inclusión de contraloría
crmp_contraloria['Alineación con subobjetivo']="Contraloría social"


crmp_contraloria_dict={}

for i in range(len(crmp_contraloria)):
    crmp_contraloria_dict.update({crmp_contraloria.iloc[i,0]: crmp_contraloria.iloc[i,2].split("/")})
```

```python
# Generamos una función que realizará el preprocesamiento al texto
def nltk_fun(programa):
    # Cambiamos mayúsculas por minúsculas
    programa= programa.lower()

    # Removemos Stopwords
    import nltk
    from nltk.corpus import stopwords
    # stop contiene la lista de stopwords
    stop = stopwords.words('spanish')
    # removemos algunas stopwords
    for i in ["estado","estados"]:
        stop.remove(i)

    # removemos stopwords del programa
    programa=" ".join(x for x in programa.split() if x not in stop)

    # Removemos puntuación
    import re
    programa=re.sub(r'[^\w\s]'," ",programa)

    # Realizamos el stemming
    from nltk.stem import SnowballStemmer

    stemmer = SnowballStemmer("spanish")

    programa=" ".join(stemmer.stem(y) if not y in excepciones_stemming else y  for y in programa.split() )

    # Eliminamos los espacios innecesarios
    programa=re.sub(' +',' ',programa)

    return programa
```

```python
# Aplicamos el método nltk_fun de preeprocesador al diccionario de categorias de los objetivos 1,2,3 y 4
for (k, v) in cat_dic_1_3.items():
    cat_dic_1_3[k]=list(map(nltk_fun,v))
# Aplicamos el método nltk_fun de preeprocesador al diccionario de categorias del objetivo 5
for (k, v) in cat_dic_4.items():
    cat_dic_4[k]=list(map(nltk_fun,v))
# Aplicamos el método nltk_fun de preeprocesador al diccionario de categorias del objetivo 5
for (k, v) in cat_dic_5.items():
    cat_dic_5[k]=list(map(nltk_fun,v))
```

```python
desc_fin=base['CICLO'].map(str)+base['RMP'].map(str)+","+base['DESC_FIN'].apply(nltk_fun)
desc_prop=base['CICLO'].map(str)+base['RMP'].map(str)+","+base['DESC_PROPOSITO'].apply(nltk_fun)
desc_pp=base['CICLO'].map(str)+base['RMP'].map(str)+","+base['DESC_PP'].apply(nltk_fun)
desc_ai=base['CICLO'].map(str)+base['RMP'].map(str)+","+base['DESC_AI'].apply(nltk_fun)
```

```python
desc_fin=desc_fin.apply(lambda x: x.split(","))
desc_prop=desc_prop.apply(lambda x: x.split(","))
desc_pp=desc_pp.apply(lambda x: x.split(","))
desc_ai=desc_ai.apply(lambda x: x.split(","))
```

```python
#********************************************************************#
#*****  Definimos algunas funciones para realizar la clasificación **#
#********************************************************************#

#******* Función que regresa un booleano si encuentra un match exacto

def match_exacto(programa, categorias):
    exacto_value=all(bool(re.findall('\\b'+word+'\\b', programa)) for word in categorias)
    return exacto_value

#******* Función que regresa un booleano si encuentra un match inicial

def match_inicio(programa, categorias):
    inicio_value=all(bool(re.findall(r'\b'+word+'\w+', programa)) for word in categorias)
    return inicio_value

#******* Función que regresa un booleano si encuentra un match exacto

def genera_alineacion(id, lista):
     contador=0
     resultado=""

     for i in lista:
         if contador==0:
             resultado=resultado+id+".p "+ i
             contador+=1
         else:
             if i=="":
                 break
             else:
                 resultado=resultado+" "+ id+".a"+str(contador)+" "+i
                 contador+=1
     return resultado

#******* Función que implementa el algoritmo de clasificación

def califica_programa(programa):
    # Obtenemos el crmp del programa
    crmp=programa[0]
    # Obtenemos el ramo
    ramo = int(re.search("([^a-zA-Z]+)",crmp[4:]).group(0))
    # Obtenemos la calificación del programa
    programa_calif=programa[1]
    # Definimos una lista que irá agregando los resultados
    calificaciones=[]
    # Definimos una lista que irá agregando las alineaciones
    alineacion=[]
    # Definimos una lista de tipo de gasto
    tipo_gasto=[]
    # Denimos una lista que tendrá la calificación parcial
    calif_parcial=[]

    # Checamos criterios de exclusión
    if crmp in crmp_exclusion:
        tipo_gasto.append("No social")
    # Checamos criterios de inclusión
    elif crmp in crmp_inclusion_dict:
        calif_parcial.append('/'.join(crmp_inclusion_dict[crmp]))
        tipo_gasto.append("Social")

    # En caso contrario, comenzamos la calificación
    else:

        for (k1, v1), (k2, v2),(k3, v3) in zip(cat_dic_1_3.items(),id_cat_dic_1_3.items() ,subobjetivo_dic_1_3.items()):

            exacto_value=match_exacto(programa_calif, v1)
            inicio_value=match_inicio(programa_calif, v1)

            if(exacto_value==False & inicio_value==False):
                pass
            else:
                alineacion.append(genera_alineacion("".join(v2),v1))
                calificaciones.append(''.join(v3))
                tipo_gasto.append("Social")
                if ''.join(v2)[0]=="1":
                    for (k1, v1), (k2, v2),(k3, v3) in zip(cat_dic_5.items(),id_cat_dic_5.items() ,subobjetivo_dic_5.items()):
                        exacto_value=match_exacto(programa_calif, v1)
                        inicio_value=match_inicio(programa_calif, v1)

                        if(exacto_value==False & inicio_value==False):
                            pass
                        else:
                            alineacion.append(genera_alineacion("".join(v2),v1))
                            calificaciones.append(''.join(v3))
                else:
                    pass
        if ramo in ramos_obj_4_list:
            for (k1, v1), (k2, v2),(k3, v3) in zip(cat_dic_4.items(),id_cat_dic_4.items() ,subobjetivo_dic_4.items()):

                exacto_value=match_exacto(programa_calif, v1)
                inicio_value=match_inicio(programa_calif, v1)

                if(exacto_value==False & inicio_value==False):
                    pass
            else:
                alineacion.append(genera_alineacion("".join(v2),v1))
                calificaciones.append(''.join(v3))
                tipo_gasto.append("Social")

    if crmp in crmp_contraloria_dict:
        calif_parcial.append('/'.join(crmp_contraloria_dict[crmp]))
        tipo_gasto.append("Social")
                
    # Si no encontró coincidencia, agregamos el tipo de gasto No social
    if not tipo_gasto:
        tipo_gasto.append("No social")


    return [' | '.join(list(set(alineacion))),'/'.join(list(set(calificaciones))),''.join(list(set(tipo_gasto))),'/'.join(list(set(calif_parcial)))]
```

```python
desc_fin_calif=pd.Series(len(desc_fin))
desc_prop_calif=pd.Series(len(desc_prop))
desc_pp_calif=pd.Series(len(desc_pp))
desc_ai_calif=pd.Series(len(desc_ai))

from tqdm import tqdm
for i in tqdm(range(len(desc_fin))):
  desc_fin_calif[i]=califica_programa(desc_fin[i])
  desc_prop_calif[i]=califica_programa(desc_prop[i])
  desc_pp_calif[i]=califica_programa(desc_pp[i])
  desc_ai_calif[i]=califica_programa(desc_ai[i])
```

```python
desc_fin_calif_df=pd.DataFrame(desc_fin_calif.values.tolist(), columns =['coincidencia', 'subobjetivo','tipo_gasto','calif_parcial'])
desc_prop_calif_df=pd.DataFrame(desc_prop_calif.values.tolist(), columns =['coincidencia', 'subobjetivo','tipo_gasto','calif_parcial'])
desc_pp_calif_df=pd.DataFrame(desc_pp_calif.values.tolist(), columns =['coincidencia', 'subobjetivo','tipo_gasto','calif_parcial'])
desc_ai_calif_df=pd.DataFrame(desc_ai_calif.values.tolist(), columns =['coincidencia', 'subobjetivo','tipo_gasto','calif_parcial'])
```

```python

base[[ '¿Qué encontró? FIN','¿Con qué subobjetivos de la política social se alinea? FIN']]=desc_fin_calif_df[['coincidencia', 'subobjetivo']]
base[[ '¿Qué encontró? PROPOSITO','¿Con qué subobjetivos de la política social se alinea? PROPOSITO']]=desc_prop_calif_df[['coincidencia', 'subobjetivo']]
base[[ '¿Qué encontró? PP','¿Con qué subobjetivos de la política social se alinea? PP']]=desc_pp_calif_df[['coincidencia', 'subobjetivo']]
base[[ '¿Qué encontró? AI','¿Con qué subobjetivos de la política social se alinea? AI']]=desc_ai_calif_df[['coincidencia', 'subobjetivo']]
```

```python

base['¿Con qué subobjetivos de la política social se alinea?']=desc_fin_calif_df['subobjetivo'].map(str)+"/"+desc_prop_calif_df['subobjetivo'].map(str)+"/"+desc_pp_calif_df['subobjetivo'].map(str)+"/"+desc_ai_calif_df['subobjetivo'].map(str)+"/"+desc_ai_calif_df['calif_parcial'].map(str)
base[['¿Con qué subobjetivos de la política social se alinea?']]=base['¿Con qué subobjetivos de la política social se alinea?'].apply(lambda x : "/".join(set(x.split('/'))))

def remueve_diagonal(palabra):
    if palabra=="":
        return palabra
    else:
        if palabra[0]=="/":
            return palabra[1:]          
        else:
            return palabra

base[['¿Con qué subobjetivos de la política social se alinea?']]=base['¿Con qué subobjetivos de la política social se alinea?'].apply(lambda x: remueve_diagonal(x))

tipo_gasto_conjunto=desc_fin_calif_df['tipo_gasto'].map(str)+","+desc_prop_calif_df['tipo_gasto'].map(str)+","+desc_pp_calif_df['tipo_gasto'].map(str)+","+desc_ai_calif_df['tipo_gasto'].map(str)

def unico(valor):
    if valor:
        return "Social"
    else:
        return "No social"

base['¿Es gasto social?']=tipo_gasto_conjunto.apply(lambda x: unico("Social" in x.split(",")))
```

```python

import xlsxwriter

# Creamos un pandas Excel writer usando la biblioteca xlsxwriter [1:]
writer = pd.ExcelWriter("GastoSocialCalificado_2020.09.10.xlsx", engine='xlsxwriter')

# Convertimos nuestro data frame a un objeto  XlsxWriter Excel
base.to_excel(writer, sheet_name='Base Calificada', startrow=1, index=False)

# Generamos los objetos xlsxwriter workbook y worksheet
workbook  = writer.book
worksheet = writer.sheets['Base Calificada']

# Definimos el formato de los encabezados

header_format = workbook.add_format({
    'bold': True,
    'text_wrap': True,
    'valign': 'top',
    'fg_color': '#D7E4BC',
    'align': 'center',
    'valign': 'vcenter',
    'border': 1})

header_format.set_font_size(10)

# Aplicamos el formato de encabezado a cada columna
for col_num, value in enumerate(base.columns.values):
    worksheet.write(1, col_num , value, header_format)

# Definimos el tamaño del alto de los encabezados
worksheet.set_row(1, 80)

# Definimos el ancho de las columnas
worksheet.set_column('A:W', 15)


#### Definimos el formato de las celdas que están combinadas
# Lista de los colores a usar
hex_color=['#b7ded2','#f6a6b2','#f7c297','#ffecb8','#90d2d8','#3b83bd']
# Lista de los nombres de las columnas combinadas
nombre_columnas=['FIN','PROPÓSITO','NOMBRE DEL PROGRAMA','ACTIVIDAD INSTITUCIONAL','ETIQUETADO FINAL']
# Rangos de cada columna combinada
columnas_excel=['L1:M1','N1:O1','P1:Q1','R1:S1','T1:U1']

for i in range(5):
    # Definimos el formato a utilizar
    merge_format = workbook.add_format({
        'bold': 1,
        'border': 1,
        'fg_color': hex_color[i],
        'align': 'center',
        'valign': 'vcenter'})

    # Combinamos las celdas
    worksheet.merge_range(columnas_excel[i], nombre_columnas[i], merge_format)


# Cerramos el Pandas Excel writer y guardamos la salida como un archivo Excel
writer.save()
```
